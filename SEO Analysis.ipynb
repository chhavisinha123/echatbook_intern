## 1. Importing library and functions

```python
import pandas as pd
import advertools as adv
pd.options.display.max_columns = None
from devtool import g_search, yt_search
```

## 2. Searching queries on Google

```python
query=["Sushant Singh Rajpoot", "butter chicken", "lenovo", "coronavirus", "immunity"]
df= g_search(query)

df.head(15)
# Here is a sample of the Google data showing the first ten columns. 
```

```python
df.info()
```

```python
print('Total number of rows:',df.shape[0], 'and columns:', df.shape[1])
```

## 2.i. Creating quick summary

- `count`: The number of times the domain appeared in the dataset. Keep in mind that the same domain might appear more than once, in the same query. As you can see below YouTube appeared 80% of the queries (4 ÷ 5 unique queries).
- `avg_rank`: The average rank that the domain appeared on, in the dataset. 
- `coverage`: The number of times the domain appeared ÷ the number of unique queries sent.

```python
summary = (df
           .groupby(['displayLink'], as_index=False)
           .agg({'rank': ['count', 'mean']})
           .sort_values(('rank', 'count'), ascending=False))
# creating a dataframe by grouping all link element with using agg function(count & mean) on rank column and also sorting according to the count of rank column.
           
summary.columns = ['DisplayLink', 'Count', 'Avg_Rank']
# inserting these list element as column name

summary['Avg_Rank'] = summary['Avg_Rank'].round(0)
# mean= 6.750000-----> 7.00

summary['DisplayLink'] = summary['DisplayLink'].str.replace('www.', '')
# www.fb.com----> fb.com


summary['coverage']=summary['Count'].div(df['searchTerms'].nunique()).mul(100).astype(str).add('%')
# coverage = (total count of that domain)/ (total queries we have searched) * 100 %

summary.head(10)

# yt ssr -4
# yt ssr - 6
# yt butterchkn - 9
# yt immunity - 9

# yt count--- > 4
# yt rank -----> (4+6+9+9)/4
# yt coverage -----> 4/5 *100%
```

**Inference:** YouTube is clearly the undisputed leader, and very far from the first competitor.

## 3. Searching queries on YouTube

```python
query=["Sushant Singh Rajpoot", "butter chicken", "lenovo", "coronavirus", "immunity"]
df2= yt_search(query)

df2.head(15)
# Here is a sample of the Youtube data showing the first ten columns. 
```

```python
df2.info()
```

```python
print('Total number of rows:',df2.shape[0], 'and columns:', df2.shape[1])
```

**Inference : ** Note the difference in the number of columns here. 85 compared to 304 on Google search. 

## 3.i. Creating quick summary

- `count`: The number of times the domain appeared in the dataset. Keep in mind that the same domain might appear more than once, in the same query.
- `avg_rank`: The average rank that the domain appeared on, in the dataset. 
- `coverage`: The number of times the domain appeared ÷ the number of unique queries sent.

```python
summary2 = (df2
           .groupby(['channelTitle'], as_index=False)
           .agg({'rank': ['count', 'mean']})
           .sort_values(('rank', 'count'), ascending=False))
# creating a dataframe by grouping all link element with using agg function(count & mean) on rank column and also sorting according to the count of rank column.
           
summary2.columns = ['Channel_Name', 'Count', 'Avg_Rank']
# inserting these list element as column name


summary2['Avg_Rank'] = summary2['Avg_Rank'].round(0)
# mean= 6.750000-----> 7.00


summary2['Coverage']=summary2['Count'].div(df2['q'].nunique()).mul(100).astype(str).add('%')
# coverage = (total count of that domain)/ (total queries we have searched) * 100 %
# df2['q'].nunique() === len(query) === 5

summary2.head(15)

# TSeries ssr -4
# TSeries ssr - 6
# TSeries butterchkn - 9
# TSeries immunity - 9

# TSeries count--- > 4
# TSeries rank -----> (4+6+9+9)/4
# TSeries coverage -----> 4/5 *100%
```

**Inference : ** Here the picture is completely different from Google search. YouTube returns five results per query, so this also has an effect on how data are distributed.

## 3.ii. Text Analysis

Things get more interesting with weighted frequency.  
Here is the word frequency for YouTube's DataFrame, counting words in the title column, with videos' number of views.

```python
# Case 1: Below is the relevant word count in Youtube video's title. After removing stopwords.

stop_words = list(adv.stopwords['english']) + ['-', '', '|', '–', '&']

ytb_title_word_freq = adv.word_frequency(text_list=df2['title'], rm_words=stop_words)

ytb_title_word_freq.head(15)
```

```python
# Case 2: Below is the relevant word count in Youtube video's description. After removing stopwords.

stop_words = list(adv.stopwords['english']) + ['-', '', '|', '–', '&']

ytb_desc_word_freq = adv.word_frequency(text_list=df2['description'],rm_words=stop_words)
# df2.'description=['ssr died in 2020', 'ssr died by suicide'......]

ytb_desc_word_freq.head(15)
```

```python
# Case 3: Below is the relevant video tags count in Youtube video. After removing stopwords.

stop_words = list(adv.stopwords['english']) + ['-', '', '|', '–', '&']

list_of_tags=[]
for i in df2['video.tags'].fillna(""):
    list_of_tags+=i

ytb_tags_word_freq = adv.word_frequency(text_list=list_of_tags , rm_words=stop_words)

ytb_tags_word_freq.head(15)
```

```python
# Case 4: Below is the relevant Hashtags count in Youtube video's description. After removing stopwords.

stop_words = list(adv.stopwords['english']) + ['-', '', '|', '–', '&']

ytb_hashtags_word_freqadv= adv.word_frequency(df2['video.description'].fillna(' '), regex=adv.regex.HASHTAG_RAW, rm_words=stop_words)

ytb_hashtags_word_freqadv.head(15)
```
